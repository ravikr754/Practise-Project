```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
x = df.drop(columns='class')
y = df['class']
```

```python
acc = 0
best_rs = 0

for i in range(1,300):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=i)
    LE = LogisticRegression()
    LE.fit(x_train,y_train)
    y_pred = LE.predict(x_test)
    mod_acc = accuracy_score(y_test,y_pred)
    if mod_acc > acc:
        acc = mod_acc
        best_rs = i
        
best_rs
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=best_rs)
LG = LogisticRegression()
LG.fit(x_train,y_train)
y_pred = LG.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print("The accuracy score for the Logistic Regression model is =",acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(min_samples_split = 10)
RFC.fit(x_train,y_train)
y_pred1 = RFC.predict(x_test)
acc1 = accuracy_score(y_test,y_pred1)
print("The accuracy score for the RandomForestClassifier is =",acc1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KN = KNeighborsClassifier()
KN.fit(x_train,y_train)
y_pred2 = KN.predict(x_test)
acc2 = accuracy_score(y_test,y_pred2)
print("The accuracy score for the KNeighborClassifier is =",acc2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier()
GBC.fit(x_train,y_train)
y_pred3 = GBC.predict(x_test)
acc3 = accuracy_score(y_test,y_pred3)
print("The accuracy score for GradientBoostingClassifier is =",acc3)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv = cross_val_score(LG,x,y,cv = 6)
cv = cv.mean()
cv1 = cross_val_score(RFC,x,y,cv = 6)
cv1 = cv1.mean()
cv2 = cross_val_score(KN,x,y,cv = 6)
cv2 = cv2.mean()
cv3 = cross_val_score(GBC,x,y,cv = 6)
cv3 = cv3.mean()
```

```python
auc_score = [acc,acc1,acc2,acc3]
cv_score = [cv,cv1,cv2,cv3]
overfit = pd.DataFrame({})
overfit['auc_score'] = auc_score
overfit['cv_score'] = cv_score
overfit['Difference'] = overfit['auc_score']-overfit['cv_score']
overfit
```

```python
params = {'n_estimators':[50,100,150,200],
         'criterion': ['entropy','gini'],
         'max_depth':[3,5,6],
         'max_features':['auto', 'sqrt', 'log2'],
         'min_samples_split':[8,10,12,14]}
gsc = GridSearchCV(RandomForestClassifier(), params, cv = 5)
gsc.fit(x_train,y_train)
```

```python
gsc.best_params_
```

```python
grand = RandomForestClassifier(bootstrap=True,
                              criterion = 'gini',
                              max_depth = 6,
                              max_features = 'sqrt',
                              n_estimators = 200)
grand.fit(x_train,y_train)
g_pred = grand.predict(x_test)
acc_g = accuracy_score(y_test,g_pred)
print("The Accuracy Score of the hyper parameter tuning is =",acc_g)
```

```python
cr = classification_report(y_test,g_pred)
print("The classification report of the model is\n",cr)
```
